```python
import numpy as np
import matplotlib.pyplot as plt

# 1.   
G = 6.67430e-11  #  
M = 5.972e24  #   (kg)
r_min = 1e6  #   (m)
r_max = 10e6  #   (m)

#  
x = np.linspace(-r_max, r_max, 100)
y = np.linspace(-r_max, r_max, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)  #  

#    (Z = r)
Phi = -G * M / np.where(Z == 0, np.nan, Z)  # Z 0  NaN 

# 2. 3D 
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Phi, cmap='viridis')

#  
ax.set_title('Gravitational Potential')
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Potential (Phi)')

plt.show()
```
