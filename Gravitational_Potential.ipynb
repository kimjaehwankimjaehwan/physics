```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 포텐셜 함수 계산
G = 6.67430e-11  # 중력 상수
M = 5.972e24  # 지구의 질량 (kg)
r_min = 1e6  # 최소 거리 (m)
r_max = 10e6  # 최대 거리 (m)

# 격자 생성
x = np.linspace(-r_max, r_max, 100)
y = np.linspace(-r_max, r_max, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)  # 거리 계산

# 포텐셜 함수 계산 (Z = r)
Phi = -G * M / np.where(Z == 0, np.nan, Z)  # Z가 0인 경우는 NaN으로 처리

# 2. 3D 시각화
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Phi, cmap='viridis')

# 축 설정
ax.set_title('Gravitational Potential')
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Potential (Phi)')

plt.show()
```


* 중심에 있는 깊은 부분이 중력 포텐셜이 큰 영역을 나타냅니다. 이는 질량이 큰 물체(예: 지구)에 의해 형성된 중력 우물(gravitational well)을 시각적으로 표현한 것입니다.
* 포텐셜 값은 중심에서 멀어질수록 감소하며, 이는 중력장의 세기가 거리의 증가에 따라 약해지는 것을 의미합니다.
