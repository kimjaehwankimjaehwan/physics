###1. Hubble-Lema√Ætre Law Formula
The Hubble-Lema√Ætre Law is expressed by the following basic formula:

v = H_0 √ó d

Where:
- v is the velocity at which a galaxy is moving away from Earth, measured in km/s.
- d is the distance to the galaxy, measured in Mpc (megaparsecs). 1 Mpc is approximately 3.26 million light-years.
- H‚ÇÄ is the Hubble constant, which represents the current rate of expansion of the universe. This value is about 70 km/s/Mpc but can vary slightly depending on the measurement method.

###2. Theoretical Background
####2.1. Redshift
The fundamental basis of the Hubble-Lema√Ætre Law is the phenomenon of redshift. Redshift refers to the lengthening of the wavelength of light, which occurs as galaxies move away. The light from these galaxies shifts towards the red end of the spectrum as their velocity increases.

Redshift formula:

z = (Œª_observed - Œª_emitted) / Œª_emitted

Where:
- z is the redshift value.
- Œª_observed is the observed wavelength of light.
- Œª_emitted is the original wavelength of the emitted light.

As galaxies move away faster, the z value increases, indicating that more distant galaxies are receding at higher velocities, in accordance with the Hubble-Lema√Ætre Law.
####2.2. Big Bang Theory and Universal Expansion
The Big Bang Theory suggests that the universe began approximately 13.8 billion years ago from a highly dense state and has been expanding ever since. The Hubble-Lema√Ætre Law provides crucial evidence for this ongoing expansion.

- Universal Expansion: According to the Hubble-Lema√Ætre Law, as the universe expands, all galaxies are moving away from each other. This means that the universe is increasing in size over time.
- Hubble Time: The age of the universe can be estimated using the Hubble constant, known as Hubble Time, calculated as T_H = 1 / H_0. This gives an approximate age of the universe around 14 billion years.
###3. Real-world Examples
####3.1. Hubble‚Äôs Early Observations
In 1929, Edwin Hubble measured the redshifts of several galaxies and discovered that they were moving away from Earth at velocities proportional to their distances. This led to the establishment of the Hubble Law, which was the first solid evidence of the universe's expansion.
####3.2. Cosmic Microwave Background (CMB)
The Cosmic Microwave Background (CMB) is the afterglow radiation from the Big Bang, emitted approximately 380,000 years after the event. The uniformity and slight temperature fluctuations in the CMB are explained by the Hubble-Lema√Ætre Law and the expansion of the universe.
####3.3. Observational Redshift Surveys
Modern astronomy uses instruments like the Hubble Space Telescope to observe redshifts of distant galaxies. These observations contribute to understanding the early state of the universe and its current expansion rate.

- For example, the Sloan Digital Sky Survey (SDSS) has measured redshifts of millions of galaxies, analyzing the large-scale structure of the universe.
- Type Ia Supernovae: These supernovae are used to measure the Hubble constant very precisely due to their consistent brightness, allowing for accurate distance and velocity comparisons of galaxies.
###4. Differences in Hubble Constant Measurements
The Hubble constant can vary slightly depending on the measurement method, which is a significant topic of debate in astronomy.

- Measurements using the Cosmic Distance Ladder method yield a value around 73 km/s/Mpc, while measurements using the Cosmic Microwave Background (CMB) yield about 67 km/s/Mpc.
- Resolving this discrepancy is a critical challenge in modern cosmology.
Conclusion
The Hubble-Lema√Ætre Law is a crucial astronomical discovery showing that the universe is expanding, and this expansion can be measured using redshift observations. This law is fundamental to our understanding of the origin, structure, and evolution of the universe, providing deep insights into its past, present, and future.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

file_path = '/content/drive/MyDrive/·ÑÜ·ÖÆ·ÜØ·ÑÖ·Öµ·Ñí·Ö°·Ü®/simulatedHubblesLawData.csv'
data = pd.read_csv(file_path)

c = 3 * 10**5

data['velocity'] = c * data['redshift']

# (ÏòàÏãú) Í≤âÎ≥¥Í∏∞ Îì±Í∏âÏúºÎ°úÎ∂ÄÌÑ∞ Í±∞Î¶¨Î•º Ï∂îÏ†ïÌï† Ïàò ÏûàÏúºÎÇò, Ïó¨Í∏∞ÏÑúÎäî Ï†ÅÏÉâÌé∏Ïù¥Î•º ÌÜµÌï¥ ÏÜçÎèÑ-Í±∞Î¶¨ Í¥ÄÍ≥ÑÎßå ÏÇ¨Ïö©
# Ïã§Ï†ú Î∂ÑÏÑùÏóêÏÑúÎäî Ï∂îÍ∞Ä Ï†ïÎ≥¥Í∞Ä ÌïÑÏöîÌï† Ïàò ÏûàÏùå
# Îç∞Ïù¥ÌÑ∞Î•º ÏÑ†Ìòï ÌöåÍ∑Ä Î∂ÑÏÑùÏóê ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎèÑÎ°ù Í±∞Î¶¨ Ï∂îÏ†ï (redshiftÎ•º ÏÉÅÎåÄÏ†ÅÏù∏ Í±∞Î¶¨Î°ú Í∞ÑÏ£º)

slope, intercept, r_value, p_value, std_err = linregress(data['redshift'], data['velocity'])

hubble_constant = slope
print(f"Í≥ÑÏÇ∞Îêú ÌóàÎ∏î ÏÉÅÏàò H_0: {hubble_constant:.2f} km/s/Mpc")
print(f"ÏÑ†Ìòï ÌöåÍ∑ÄÏùò Í≤∞Ï†ïÍ≥ÑÏàò (R^2): {r_value**2:.3f}")

plt.figure(figsize=(10, 6))
plt.scatter(data['redshift'], data['velocity'], color='blue', label='Observed Data')
plt.plot(data['redshift'], intercept + slope*data['redshift'], 'r', label=f'Fit: H_0 = {hubble_constant:.2f} km/s/Mpc')
plt.xlabel('Redshift (z)')
plt.ylabel('Velocity (km/s)')
plt.title('Hubble-Lema√Ætre Law')
plt.legend()
plt.grid(True)
plt.show()
```
